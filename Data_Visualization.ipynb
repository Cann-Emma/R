# DATA VISAUALIZATION

```r
#Installing needed packages
install.packages('RColorBrewer')
```

```r
# Loading packages
library(ggplot2)
library(dplyr)
library(RColorBrewer)
```

```r
# Importing Dataset and Getting an idea of the variables in it
RC= read.csv('RollerCoasterData.csv')
glimpse(RC)
```

```r
#Bubble Plot: Using more than two variables on a graph
RC%>%ggplot(aes(x= Height, y= Speed, fill= Type, size= age))+geom_point(alpha= 0.5, shape= 21)+scale_size(range= c(0.1, 13))
```

```r
# Percent Stacked Bar Graph and Changing the class/data type of variables
RC1= RC%>%mutate(Inv= as.character(Inversions))
RC1%>%ggplot(aes(x= Inv, fill= Design))+geom_bar(position = "fill")
```

```r
ggplot(RC)+geom_histogram(aes(x= age), colour= 'black', fill= 'turquoise', binwidth = 7)+ggtitle('Distribution of RollerCoaster by Age')+xlab('Age of RollerCoaster')+ylab('Number of RollerCoaster')+theme_bw()+theme(panel.grid= element_blank(), plot.title= element_text(hjust = 0.5))
```

```r
RC%>%ggplot(aes(x= Design))+geom_bar(fill= brewer.pal(7, "Set1"))
```

```r
RC%>%ggplot(aes(x=State, y= age, fill= Type))+geom_boxplot()+ theme(axis.text.x = element_text(angle= 90))
```

```r
RC%>%ggplot(aes(x=Height, y= Speed, color= Type))+geom_point()+scale_color_manual(values = c("azure3","burlywood3"))+theme(panel.grid= element_blank())
```
